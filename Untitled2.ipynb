Movie Booking System

```python
print("welcome to movie booking system")
available_seats= 5
password = "abcd"
```

```python
print("what do you want to do")
print("1. book tickets")
print("2. check available seats")
print("3. admin add seats")
print("4. exit")
```

```python
while True:
  choice = int(input("whats your choice"))

  if choice == 1:
      print("tickets booked")
  elif choice == 2:
      print("available seats" ,available_seats )
  elif choice == 3:
      admin_password = input("enter your password to add seat")
      if admin_password == password:
        new_seats = int(input("how many seats do you want to add"))
        available_seats = available_seats + new_seats
        print("success")
        print("current available seats" , available_seats)
      else :
        print("wrong password")
  elif choice == 4:
      print("exiting")
      break
  else :
      print("not a valid choice ")
```
